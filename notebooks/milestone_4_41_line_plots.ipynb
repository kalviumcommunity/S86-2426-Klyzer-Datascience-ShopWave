# Milestone 4.41: Identifying Trends Over Time Using Line Plots

**Objective:** Visualize and interpret how values evolve over time using line plots.

**Key Learning:**
- Time must be treated as ordered data
- Line plots show continuity across a timeline
- Trend direction and volatility become visually clear
- Spikes and drops can reveal important signals

---

## Why Time-Based Visualization Matters

Common beginner issues:
- Treating time data as unordered rows
- Missing long-term trend by looking only at snapshots
- Overreacting to one or two points
- Ignoring temporal context when interpreting values

Professional habit: convert to datetime, sort by time, then plot.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

print('Pandas version:', pd.__version__)
```

---

## Part 1: Load Data and Prepare Time Column

```python
sales_df = pd.read_csv('../data/raw/sales_data.csv')

print('Raw data preview:')
print(sales_df.head())
print('\nData types before conversion:')
print(sales_df.dtypes)
```

```python
# Convert date column to datetime and sort chronologically
sales_df['OrderDate'] = pd.to_datetime(sales_df['OrderDate'])
sales_df = sales_df.sort_values('OrderDate')

print('Data types after conversion:')
print(sales_df.dtypes)
print('\nTime-ordered preview:')
print(sales_df[['OrderDate', 'Quantity', 'Price']].head())
```

Time ordering check complete. We can now analyze trends correctly.

---

## Part 2: Create Line Plots

```python
# Line plot for Price over time
plt.figure(figsize=(9, 4))
plt.plot(sales_df['OrderDate'], sales_df['Price'], marker='o')
plt.title('Price Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Optional second line plot for Quantity
plt.figure(figsize=(9, 4))
plt.plot(sales_df['OrderDate'], sales_df['Quantity'], marker='o', color='tab:orange')
plt.title('Quantity Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Interpretation prompts:
- Is the overall trend upward, downward, or mixed?
- Are there periods that look stable?
- Do some days show sudden changes?

---

## Part 3: Identify Trend Direction

```python
price_start = sales_df['Price'].iloc[0]
price_end = sales_df['Price'].iloc[-1]
price_change_total = price_end - price_start

print('Price start:', price_start)
print('Price end:', price_end)
print('Total change:', price_change_total)

if price_change_total > 0:
    print('Overall direction: Upward trend (based on endpoints).')
elif price_change_total < 0:
    print('Overall direction: Downward trend (based on endpoints).')
else:
    print('Overall direction: Flat trend (based on endpoints).')
```

Endpoint comparison is useful, but always inspect the full line shape before concluding.

---

## Part 4: Spot Spikes, Drops, and Volatility

```python
sales_df['price_change'] = sales_df['Price'].diff()

largest_rises = sales_df.nlargest(2, 'price_change')[['OrderDate', 'Price', 'price_change']]
largest_drops = sales_df.nsmallest(2, 'price_change')[['OrderDate', 'Price', 'price_change']]

print('Largest rises:')
print(largest_rises)
print('\nLargest drops:')
print(largest_drops)
```

```python
# Visual highlight of notable daily changes
plt.figure(figsize=(9, 4))
plt.plot(sales_df['OrderDate'], sales_df['price_change'], marker='o', color='tab:red')
plt.axhline(0, linestyle='--', color='black')
plt.title('Day-to-Day Price Change Over Time')
plt.xlabel('Order Date')
plt.ylabel('Price Change')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Spikes and drops are not automatically errors. Treat them as prompts for deeper investigation.
