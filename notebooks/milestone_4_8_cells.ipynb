**Purpose:** This notebook demonstrates the fundamental difference between Code cells and Markdown cells in Jupyter Notebooks.

**Learning Objectives:**
- Understand when to use Code cells vs Markdown cells
- Create structured, readable notebooks

---

## Part 1: What Are Markdown Cells?

**Markdown cells are for explanation, not execution.**

They allow you to:
- Add headings and titles
- Write narrative text to explain your analysis
- Create lists and bullet points
- Document your reasoning and methodology

**This is a Markdown cell!** It contains formatted text, not executable code.

## Part 2: What Are Code Cells?

The next cell is a **Code cell**. It will execute Python code when you run it (Shift + Enter).

Code cells:
- Execute Python statements
- Display output below the cell
- Perform computations and operations
- Should contain logic, not lengthy explanations

````python
# This is a Code cell - it executes Python code
print("Hello! This is executable code.")
print("Code cells run computations and show results.")
````

**Did you notice the difference?**

- The Markdown cell above displayed formatted text
- The Code cell executed Python and showed output

---

## Part 3: Example - Simple Math Operation

Below, I'll demonstrate a simple calculation. **Notice how Markdown explains WHAT we're doing, and Code DOES it.**

### Example: Calculate the area of a rectangle

**Objective:** Calculate the area of a rectangle with:
- Length = 10 units
- Width = 5 units

**Formula:** Area = Length Ã— Width

````python
# Define dimensions
length = 10
width = 5

# Calculate area
area = length * width

# Display result
print(f"The area of the rectangle is: {area} square units")
````

**Result Interpretation:**

The calculation shows that a 10Ã—5 rectangle has an area of 50 square units. This demonstrates how Code performs the calculation while Markdown provides context before and after.

---

## Part 4: Example - Working with Lists

Let's create a list of numbers and calculate their sum and average.

````python
# Create a list of numbers
numbers = [12, 25, 18, 30, 22]

# Calculate sum and average
total = sum(numbers)
average = total / len(numbers)

# Display results
print(f"Numbers: {numbers}")
print(f"Sum: {total}")
print(f"Average: {average}")
````

**Observation:**

The average of the five numbers is 21.4. Notice how:
- The **Markdown cell** above explained what we're doing
- The **Code cell** performed the actual calculation
- This **Markdown cell** interprets the results

This separation makes notebooks easy to read and understand!

## Part 5: When to Use Code vs Markdown

### Use **Code Cells** when you need to:
- Execute Python statements
- Perform calculations or operations
- Import libraries
- Load or process data
- Create visualizations
- Run any computational logic

### Use **Markdown Cells** when you need to:
- Explain the purpose of your notebook
- Describe what a section of code will do
- Interpret results and findings
- Document methodology or reasoning
- Create section headers and structure
- Add context for reviewers or teammates

## Part 7: Example of Poor vs Good Notebook Structure

### âŒ Poor Structure (avoid this):
```python

# Price is 25 and quantity is 100
# The result shows total revenue
```

**Problem:** Everything is inside code comments. Hard to read, unprofessional.

### âœ… Good Structure (use this):

> - Price per unit: $25
> - Quantity sold: 100 units

**Code Cell:**
```python

quantity = 100
revenue = price * quantity
print(f"Total revenue: ${revenue}")
```

**Markdown Cell:**
> **Result:** Total revenue is $2,500. This represents strong sales performance.

**Why it's better:** Clear separation of explanation and execution. Professional and readable.
