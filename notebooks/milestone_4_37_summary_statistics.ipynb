# Milestone 4.37: Computing Basic Summary Statistics for Individual Columns

**Objective:** Build strong EDA fundamentals by computing and interpreting summary statistics for individual numeric columns in a Pandas DataFrame.

**Key Learning:**
- Summary statistics provide a quick quantitative snapshot
- Central tendency: mean and median
- Spread: min, max, standard deviation, variance
- Count helps verify usable observations
- Interpretation matters more than memorizing formulas

---

## Why This Matters

**Common beginner mistakes:**
- Jumping into analysis without understanding numeric patterns
- Trusting averages without checking spread
- Ignoring outliers that shift the mean
- Making assumptions from raw rows only

**Professional habit:** Always compute summary statistics before deeper analysis.

```python
# Import library
import pandas as pd

print('Pandas version:', pd.__version__)
```

---

## Part 1: Load Data and Inspect Numeric Columns

```python
# Load sales data
sales_df = pd.read_csv('../data/raw/sales_data.csv')

print('Sales data preview:')
print(sales_df.head())
print('\nShape:', sales_df.shape)
```

```python
# Identify numeric columns
numeric_cols = sales_df.select_dtypes(include='number').columns.tolist()
print('Numeric columns:', numeric_cols)
```

**Observation:** Numeric columns are where mean, median, and spread metrics are meaningful.

---

## Part 2: Compute Statistics for a Single Column

### Focus Column: Quantity

```python
# Select a single numeric column
quantity = sales_df['Quantity']

print('Quantity values:')
print(quantity.tolist())
```

```python
# Basic summary statistics for Quantity
qty_stats = quantity.agg(['count', 'mean', 'median', 'min', 'max', 'std', 'var'])
print('Summary statistics for Quantity:')
print(qty_stats)
```

### Interpreting Quantity Statistics

- `count`: Number of usable values
- `mean`: Average quantity per order
- `median`: Middle value (less sensitive to outliers)
- `min` / `max`: Lowest and highest observed quantities
- `std` and `var`: How spread out the values are

---

## Part 3: Mean vs Median (Outlier Sensitivity)

Outliers usually affect mean more than median.

```python
print('Original mean:', quantity.mean())
print('Original median:', quantity.median())
```

```python
# Add one extreme value to demonstrate outlier impact
quantity_with_outlier = pd.concat([quantity, pd.Series([50])], ignore_index=True)

print('With outlier mean:', quantity_with_outlier.mean())
print('With outlier median:', quantity_with_outlier.median())
```

**Insight:** If mean changes a lot but median changes little, the distribution likely has skew or unusual values.

---

## Part 4: Compare Statistics Across Columns

Now compare `Quantity` and `Price`.

```python
comparison = sales_df[['Quantity', 'Price']].agg(['count', 'mean', 'median', 'min', 'max', 'std', 'var'])
print('Comparison of Quantity vs Price:')
print(comparison)
```

```python
# Range comparison helps explain spread
qty_range = sales_df['Quantity'].max() - sales_df['Quantity'].min()
price_range = sales_df['Price'].max() - sales_df['Price'].min()

print(f'Quantity range: {qty_range}')
print(f'Price range: {price_range:.2f}')
```

### Interpretation Prompts

- Which column has larger variability?
- Is mean close to median in each column?
- Do min/max suggest possible unusual observations?
- What should you investigate next before modeling?

---

## Part 5: Build a Reusable Summary Function

```python
def basic_summary(series):
    return series.agg(['count', 'mean', 'median', 'min', 'max', 'std', 'var'])

print('Summary for Quantity:')
print(basic_summary(sales_df['Quantity']))

print('\nSummary for Price:')
print(basic_summary(sales_df['Price']))
```
