## Part 1: Understanding CSV Files

### What is a CSV File?

**CSV** = **C**omma-**S**eparated **V**alues

A plain text file format where:
- Each line represents a **row**
- Values in a row are separated by **commas** (or other delimiters)
- First row usually contains **column names** (header)
- Remaining rows contain **data**

**Real-World Analogy:**
- CSV = Plain text version of a spreadsheet
- Excel/Google Sheets can save as CSV
- Universal format for data sharing

### Example CSV Structure

**products.csv:**
```
Product,Price,Stock,Category        ← Header row (column names)
Laptop,999.99,15,Electronics        ← Data row 1
Mouse,29.99,150,Electronics         ← Data row 2
Keyboard,79.99,80,Electronics       ← Data row 3
```

**Interpretation:**
- **Delimiter:** Comma (`,`)
- **Columns:** Product, Price, Stock, Category
- **Rows:** 3 data rows (after header)
- **Structure:** 4 columns × 3 rows

### CSV vs Spreadsheet

| Aspect | CSV File | Excel Spreadsheet |
|--------|----------|-------------------|
| **Format** | Plain text | Binary (proprietary) |
| **Columns** | Separated by delimiter | Cell-based |
| **Rows** | One per line | Cell-based |
| **Formulas** | No | Yes |
| **Formatting** | No | Yes (colors, fonts) |
| **Size** | Small, compressed | Large |
| **Compatibility** | Universal | Excel-dependent |
| **Use in DS** | Primary format | Occasional |

**Why CSV dominates Data Science:**
- Simple, text-based
- Works everywhere (Python, R, SQL, Excel)
- Easy to version control (Git)
- Fast to load and parse
- No software dependencies

## Part 2: Loading CSV Files into Pandas

### The Core Function: `pd.read_csv()`

**Purpose:** Load CSV file → Create DataFrame

**Basic Syntax:**
```python
df = pd.read_csv('path/to/file.csv')
```

**What Happens:**
1. Pandas opens the file
2. Reads the first row → column names
3. Reads remaining rows → data
4. Infers data types for each column
5. Creates DataFrame with index 0, 1, 2...

````python
# Import Pandas
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
````

### Loading a Standard CSV File

**Scenario:** Load products data with standard comma delimiter and header row.

````python
# Load products.csv
products_df = pd.read_csv('../data/raw/products.csv')

print("✓ CSV file loaded successfully!")
print("\nDataFrame Preview:")
print(products_df)
````

**Observations:**
- First row (Product, Price, Stock, Category) → Column names
- Data rows → DataFrame rows with index 0, 1, 2, 3, 4
- Commas → Separate columns
- Pandas automatically infers data types

### Understanding File Paths

**Relative Paths (Recommended for Projects):**
```python
df = pd.read_csv('../data/raw/file.csv')     # Go up one level, then into data/raw
df = pd.read_csv('data/file.csv')            # From current directory
```

**Absolute Paths:**
```python
df = pd.read_csv('D:/Data-Science/data.csv')  # Full path (Windows)
df = pd.read_csv('/home/user/data.csv')       # Full path (Linux/Mac)
```

**Best Practice:**
- Use **relative paths** for portability
- Organize data files in a `data/` folder
- Keep notebooks in a `notebooks/` folder

## Part 3: Inspecting Loaded Data

### **Golden Rule:** ALWAYS inspect data after loading!

**Why?**
- Verify correct file was loaded
- Confirm column names match expectations
- Check data types are appropriate
- Detect missing values early
- Catch loading errors before analysis

**Think of it like:** You wouldn't start cooking without checking the recipe ingredients—don't analyze data without checking what loaded.

### Method 1: `.head()` - Preview First Rows

````python
# Preview first 3 rows
print("First 3 rows:")
print(products_df.head(3))

print("\nFirst 5 rows (default):")
print(products_df.head())
````

**Use `.head()` to:**
- Quickly view sample data
- Check if data looks reasonable
- Verify column names are correct

### Method 2: `.tail()` - Preview Last Rows

````python
# Preview last 2 rows
print("Last 2 rows:")
print(products_df.tail(2))
````

**Use `.tail()` to:**
- Check end of dataset
- Verify file loaded completely
- Look for unexpected patterns at the end

### Method 3: `.info()` - Comprehensive Overview

````python
# Get comprehensive information
print("DataFrame Information:")
products_df.info()
````

**What `.info()` tells you:**
- Number of rows and columns
- Column names
- Non-null counts (helps detect missing data)
- Data types (int64, float64, object)
- Memory usage

**This is one of the MOST important inspection methods!**

### Method 4: `.shape` - Dimensions

````python
# Check dimensions
print(f"Shape: {products_df.shape}")
print(f"Rows: {products_df.shape[0]}")
print(f"Columns: {products_df.shape[1]}")
````

**Use `.shape` to:**
- Verify expected number of rows
- Confirm all columns loaded
- Quick size check

### Method 5: `.columns` - Column Names

````python
# View all column names
print("Column names:")
print(products_df.columns.tolist())

print("\nNumber of columns:", len(products_df.columns))
````

**Use `.columns` to:**
- Verify column names match documentation
- Check for unexpected spaces or typos
- Confirm all expected columns are present

### Method 6: `.describe()` - Statistical Summary

````python
# Get statistical summary for numeric columns
print("Statistical Summary:")
print(products_df.describe())
````

**What `.describe()` shows:**
- count: Number of non-null values
- mean: Average
- std: Standard deviation
- min/max: Range
- 25%, 50%, 75%: Quartiles

**Use `.describe()` to:**
- Detect unrealistic values (e.g., negative prices)
- Understand data distribution
- Spot outliers

### Method 7: Missing Data Check

````python
# Check for missing values
print("Missing values per column:")
print(products_df.isnull().sum())

print("\nTotal missing values:", products_df.isnull().sum().sum())
````

**Why check missing data?**
- Missing data breaks many operations
- Need to decide how to handle (drop, fill, etc.)
- Better to know early than discover during analysis

### Complete Inspection Workflow

**Recommended inspection sequence after loading ANY CSV:**

````python
# Complete inspection workflow
print("=" * 60)
print("COMPLETE DATA INSPECTION")
print("=" * 60)

print("\n1. SHAPE:")
print(f"   {products_df.shape[0]} rows × {products_df.shape[1]} columns")

print("\n2. COLUMNS:")
print(f"   {products_df.columns.tolist()}")

print("\n3. PREVIEW (first 3 rows):")
print(products_df.head(3))

print("\n4. DATA TYPES & NULL COUNTS:")
products_df.info()

print("\n5. MISSING VALUES:")
print(products_df.isnull().sum())

print("\n6. STATISTICS:")
print(products_df.describe())

print("\n" + "=" * 60)
print("✓ INSPECTION COMPLETE - Data ready for analysis")
print("=" * 60)
````

## Part 4: Recognizing Common Loading Issues

### Issue 1: Wrong Delimiter

**Problem:** CSV uses semicolons (`;`) instead of commas.

**Example:** `sales_semicolon.csv`

````python
# INCORRECT: Using default comma delimiter
print("Loading with WRONG delimiter (comma):")
wrong_df = pd.read_csv('../data/raw/sales_semicolon.csv')
print(wrong_df.head())
print(f"\nColumns: {wrong_df.columns.tolist()}")
print("⚠️ Notice: Everything loaded into ONE column!")
````

````python
# CORRECT: Specify semicolon delimiter
print("\nLoading with CORRECT delimiter (semicolon):")
sales_df = pd.read_csv('../data/raw/sales_semicolon.csv', sep=';')
print(sales_df.head())
print(f"\nColumns: {sales_df.columns.tolist()}")
print("✓ Now data is properly separated!")
````

**Solution:** Use `sep=';'` parameter

**How to detect:**
- Only 1 column after loading
- Column name contains delimiters
- Values look like "value1;value2;value3"

### Issue 2: No Header Row

**Problem:** CSV has no column names (all rows are data).

**Example:** `data_no_header.csv`

````python
# INCORRECT: Pandas assumes first row is header
print("Loading WITHOUT specifying no header:")
wrong_df = pd.read_csv('../data/raw/data_no_header.csv')
print(wrong_df)
print("\n⚠️ Notice: First data row became column names!")
print(f"Columns: {wrong_df.columns.tolist()}")
````

````python
# CORRECT: Specify no header
print("\nLoading WITH header=None:")
employees_df = pd.read_csv('../data/raw/data_no_header.csv', header=None)
print(employees_df)
print("\n✓ Now all rows are treated as data!")
print(f"Columns: {employees_df.columns.tolist()}")
````

````python
# BETTER: Provide custom column names
print("\nLoading with CUSTOM column names:")
employees_df = pd.read_csv(
    '../data/raw/data_no_header.csv', 
    header=None,
    names=['Name', 'Department', 'Salary']
)
print(employees_df)
print("\n✓ Data loaded with meaningful column names!")
````

**Solution:** Use `header=None` and optionally `names=[...]`

**How to detect:**
- Column names look like data values
- One less row than expected
- First row missing from data

### Issue 3: File Not Found

**Problem:** Incorrect file path.

````python
# This will raise FileNotFoundError
try:
    df = pd.read_csv('nonexistent_file.csv')
except FileNotFoundError as e:
    print("❌ ERROR:", e)
    print("\n💡 How to fix:")
    print("   - Check file path is correct")
    print("   - Use relative paths: '../data/raw/file.csv'")
    print("   - Verify file exists in that location")
    print("   - Check for typos in filename")
````

### Issue 4: Extra Whitespace in Column Names

**Problem:** Column names have leading/trailing spaces.

````python
# Simulate data with spaces
import io

csv_with_spaces = """ Product , Price , Stock 
Laptop,999,15
Mouse,29,150
"""

df_spaces = pd.read_csv(io.StringIO(csv_with_spaces))
print("Column names with spaces:")
print(repr(df_spaces.columns.tolist()))
print("\n⚠️ Notice the extra spaces before/after names")

# Accessing columns becomes difficult
try:
    print(df_spaces['Price'])  # This fails
except KeyError:
    print("\n❌ Cannot access 'Price' - column name is actually ' Price '")
````

````python
# Fix: Strip whitespace from column names
df_spaces.columns = df_spaces.columns.str.strip()
print("After stripping whitespace:")
print(repr(df_spaces.columns.tolist()))
print("\n✓ Now column access works:")
print(df_spaces['Price'])
````

**Solution:** Always strip whitespace:
```python
df.columns = df.columns.str.strip()
```

### Issue 5: Incorrect Data Types

**Problem:** Pandas infers wrong data type.

````python
# Simulate data with mixed types
csv_mixed = """ID,Age,Salary
1,25,50000
2,thirty,60000
3,28,70000
"""

df_mixed = pd.read_csv(io.StringIO(csv_mixed))
print("Data types:")
print(df_mixed.dtypes)
print("\n⚠️ Notice: Age is 'object' (string) because of 'thirty'")
print("\nDataFrame:")
print(df_mixed)
````

**How to detect:**
- Use `.dtypes` to check types
- Numeric column shows as 'object' → likely has text values
- Use `.info()` to see all types at once

**Solution:**
- Clean data before loading, or
- Convert types after loading with `pd.to_numeric()`, `astype()`, etc.

### Summary of Common Issues

| Issue | Symptom | Solution |
|-------|---------|----------|
| **Wrong delimiter** | All data in one column | `sep=';'` or `sep='\t'` |
| **No header** | First data row as columns | `header=None, names=[...]` |
| **File not found** | FileNotFoundError | Check path, use relative paths |
| **Extra spaces** | Column access fails | `df.columns = df.columns.str.strip()` |
| **Wrong data types** | Numbers as 'object' | Check .dtypes, clean data |
| **Encoding issues** | Strange characters | `encoding='utf-8'` or `'latin1'` |
| **Skip rows** | Metadata/notes in file | `skiprows=N` |

**Prevention:** ALWAYS inspect data after loading with `.info()` and `.head()`

## Part 6: Best Practices for CSV Loading

### The CSV Loading Checklist

**Before Loading:**
- [ ] Know where the file is located
- [ ] Check if file has header row
- [ ] Know the delimiter (comma, semicolon, tab)
- [ ] Understand expected columns and row count

**While Loading:**
```python
# Basic load
df = pd.read_csv('file.csv')

# With parameters (as needed)
df = pd.read_csv(
    'file.csv',
    sep=',',           # Delimiter
    header=0,          # Which row is header (0 = first row)
    names=None,        # Custom column names
    skiprows=0,        # Skip N rows at start
    index_col=None,    # Use a column as index
    usecols=None,      # Load only specific columns
    dtype=None         # Specify data types
)
```

**After Loading:**
```python
# Immediate inspection (DO NOT SKIP!)
print(df.shape)           # Verify dimensions
print(df.columns.tolist()) # Check column names
print(df.head())          # Preview data
df.info()                 # Check types and nulls
print(df.isnull().sum())  # Count missing values
```

### Essential Loading Parameters

**Most commonly used `read_csv()` parameters:**

| Parameter | Purpose | Example |
|-----------|---------|----------|
| `sep` | Specify delimiter | `sep=';'` for semicolons |
| `header` | Which row is header | `header=None` if no header |
| `names` | Custom column names | `names=['A', 'B', 'C']` |
| `skiprows` | Skip rows at start | `skiprows=2` skip first 2 rows |
| `usecols` | Load specific columns | `usecols=['Name', 'Age']` |
| `index_col` | Use column as index | `index_col='ID'` |
| `dtype` | Specify data types | `dtype={'Age': int}` |
| `na_values` | Treat as missing | `na_values=['', 'N/A']` |
| `encoding` | Character encoding | `encoding='utf-8'` |
| `nrows` | Load only N rows | `nrows=1000` for testing |
