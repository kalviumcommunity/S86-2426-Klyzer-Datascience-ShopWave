# Milestone 4.40: Visualizing Data Distributions Using Boxplots

**Objective:** Use boxplots to summarize and compare numeric distributions, including center, spread, and potential outliers.

**Key Learning:**
- Boxplots summarize distributions compactly
- Median and quartiles show central tendency and spread
- IQR captures middle 50% variability
- Points beyond whiskers can indicate potential outliers

---

## Why Boxplots Matter

Common beginner issues:
- Missing outliers when using only averages
- Difficulty comparing spread across columns
- Using only histograms and missing quartile information
- Misinterpreting variability

Professional habit: combine boxplots with summary statistics for stronger EDA.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

print('Pandas version:', pd.__version__)
```

---

## Part 1: Load Data and Identify Numeric Columns

```python
sales_df = pd.read_csv('../data/raw/sales_data.csv')
numeric_cols = sales_df.select_dtypes(include='number').columns.tolist()

print('Numeric columns:', numeric_cols)
print('\nData preview:')
print(sales_df.head())
```

Boxplot components:
- Center line in box: median (Q2)
- Bottom/top of box: Q1 and Q3
- Box height: IQR (Q3 - Q1)
- Whiskers: non-outlier range
- Points beyond whiskers: potential outliers

---

## Part 2: Create a Boxplot for a Single Column

```python
plt.figure(figsize=(6, 4))
plt.boxplot(sales_df['Quantity'], vert=True, patch_artist=True)
plt.title('Boxplot of Quantity')
plt.ylabel('Quantity')
plt.tight_layout()
plt.show()
```

```python
print('Quantity summary:')
print(sales_df['Quantity'].describe())
```

Interpretation prompts:
- Where is the median located?
- Is the box (IQR) narrow or wide?
- Are any points shown outside whiskers?

---

## Part 3: Compute Quartiles and IQR Explicitly

```python
q1 = sales_df['Quantity'].quantile(0.25)
median = sales_df['Quantity'].quantile(0.50)
q3 = sales_df['Quantity'].quantile(0.75)
iqr = q3 - q1

print('Quantity quartiles and IQR:')
print('Q1:', q1)
print('Median:', median)
print('Q3:', q3)
print('IQR:', iqr)
```

```python
# Classical IQR outlier thresholds
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

outliers = sales_df[(sales_df['Quantity'] < lower_fence) | (sales_df['Quantity'] > upper_fence)]

print('Lower fence:', lower_fence)
print('Upper fence:', upper_fence)
print('Potential outliers in Quantity:')
print(outliers[['OrderID', 'Quantity', 'Price']])
```

Outliers are signals for investigation, not automatic errors.

---

## Part 4: Compare Boxplots Across Columns

```python
plt.figure(figsize=(8, 4))
plt.boxplot(
    [sales_df['Quantity'], sales_df['Price']],
    labels=['Quantity', 'Price'],
    patch_artist=True
)
plt.title('Boxplot Comparison: Quantity vs Price')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```

```python
comparison = sales_df[['Quantity', 'Price']].describe().T
comparison['IQR'] = comparison['75%'] - comparison['25%']
comparison['range'] = comparison['max'] - comparison['min']

print('Comparison stats to support boxplot interpretation:')
print(comparison[['50%', '25%', '75%', 'IQR', 'min', 'max', 'range', 'std']])
```

Comparison prompts:
- Which column has a higher median?
- Which column has wider IQR (middle spread)?
- Which column appears to have more outlier behavior?

---

## Part 5: Optional Boxplots for All Numeric Columns

```python
plt.figure(figsize=(9, 4))
sales_df[numeric_cols].boxplot()
plt.title('Boxplots for All Numeric Columns')
plt.tight_layout()
plt.show()
```
