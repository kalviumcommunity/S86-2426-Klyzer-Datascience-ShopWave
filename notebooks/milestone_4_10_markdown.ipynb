## Part 1: Understanding Markdown

**What is Markdown?**

Markdown is a lightweight markup language that lets you format text using simple syntax. In Jupyter, Markdown cells transform plain text into formatted documentation.

**Why use Markdown in notebooks?**
- Document your reasoning and methodology
- Create structure with headings and sections
- Explain code before and after execution
- Make notebooks readable for reviewers and teammates

## Part 2: Writing Headings

Headings structure your notebook and make it easy to navigate.

### Heading Syntax

Use `#` symbols to create headings:
- `# Heading 1` - Top-level heading (largest)
- `## Heading 2` - Major section
- `### Heading 3` - Subsection
- `#### Heading 4` - Sub-subsection
- `##### Heading 5` - Smaller heading
- `###### Heading 6` - Smallest heading

### Heading Best Practices

âœ… **DO:**
- Use H1 (`#`) for notebook title
- Use H2 (`##`) for major sections
- Use H3 (`###`) for subsections
- Keep headings concise and descriptive
- Maintain logical hierarchy

âŒ **DON'T:**
- Skip heading levels (H1 â†’ H4)
- Use headings for emphasis (use bold instead)
- Make headings too long or vague

## Part 9: Code Blocks

Use code blocks to display multi-line code snippets **without executing them**.

### Syntax for Code Blocks

Use triple backticks with optional language specification:

````
```python
# Your code here
```
````

### Example: Python Code Block

```python
# Import libraries
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('data.csv')

# Calculate mean
average = df['sales'].mean()
print(f"Average sales: {average}")
```

**Note:** This code is displayed but not executed. It's just for documentation.

### Example: SQL Code Block

```sql
SELECT customer_id, COUNT(*) as order_count
FROM orders
WHERE order_date >= '2024-01-01'
GROUP BY customer_id
ORDER BY order_count DESC;
```

### When to Use Code Blocks

âœ… Documenting error messages

âœ… Displaying configuration files

âœ… Explaining code structure

## Part 10: Combining Markdown and Code Cells

Professional notebooks alternate between Markdown (explanation) and Code (execution).

### Example Workflow

Below is a demonstration of proper Markdown + Code combination.

---

### Step 1: Setup

**Objective:** Import required libraries for data analysis.

**Libraries needed:**
- `math` - For mathematical operations
- `statistics` - For statistical calculations

`````python
# Import libraries
import math
import statistics

print("âœ“ Libraries imported successfully")
`````

### Step 2: Data Preparation

**Objective:** Create a sample dataset of student scores.

**Dataset details:**
- Sample size: 10 students
- Score range: 0-100
- Variable name: `scores`

`````python
# Define student scores
scores = [85, 92, 78, 95, 88, 76, 91, 83, 89, 94]

print(f"Number of students: {len(scores)}")
print(f"Scores: {scores}")
`````

### Step 3: Calculate Statistics

**Objective:** Compute basic statistical measures.

**Metrics to calculate:**
1. Mean (average score)
2. Median (middle value)
3. Standard deviation (spread of scores)

`````python
# Calculate statistics
mean_score = statistics.mean(scores)
median_score = statistics.median(scores)
stdev_score = statistics.stdev(scores)

# Display results
print("=" * 40)
print("STATISTICAL SUMMARY")
print("=" * 40)
print(f"Mean:              {mean_score:.2f}")
print(f"Median:            {median_score:.2f}")
print(f"Standard Deviation: {stdev_score:.2f}")
print("=" * 40)
`````

### Step 4: Interpretation

**Results:**
- **Mean score:** The average performance is around 87-88 points
- **Median score:** The middle score is similar to the mean, indicating relatively symmetric distribution
- **Standard deviation:** The spread is approximately 6 points, showing moderate variability

**Conclusion:**
The class performed well overall with consistent scores and minimal outliers.

## Part 12: Markdown Best Practices

### âœ… DO:

1. **Use headings to structure your notebook**
   - Makes navigation easy
   - Creates logical flow

2. **Write Markdown BEFORE code cells**
   - Explain what the code will do
   - Document assumptions and methodology

3. **Write Markdown AFTER code cells**
   - Interpret results
   - Draw conclusions

4. **Use lists for clarity**
   - Steps in a process
   - Requirements or assumptions
   - Results or findings

5. **Format code references with inline code**
   - Variable names: `x`, `data_frame`
   - Function names: `calculate()`, `process_data()`

6. **Keep Markdown concise**
   - Short paragraphs
   - Clear language
   - Focused explanations

### âŒ DON'T:

1. **Don't put explanations only in code comments**
   - Use Markdown cells for narrative
   - Keep code comments brief

2. **Don't skip documentation**
   - Every major step needs context
   - Results need interpretation

3. **Don't use inconsistent heading levels**
   - Follow logical hierarchy
   - Don't skip levels (H1 â†’ H4)

4. **Don't write overly long Markdown cells**
   - Break into smaller sections
   - Use headings to organize

5. **Don't forget to render Markdown**
   - Always press Shift + Enter to see formatted output
   - Check how it looks rendered
