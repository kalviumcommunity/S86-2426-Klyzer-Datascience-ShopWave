# Milestone 4.38: Comparing Distributions Across Multiple Columns

**Objective:** Compare distributions across multiple numeric columns to understand how variables differ in center, spread, and behavior.

**Key Learning:**
- Distribution comparison provides context that single-column analysis misses
- Mean and median compare central tendency
- Range, standard deviation, and variance compare spread
- Pattern detection supports better EDA decisions

---

## Why Comparison Matters

Common pitfalls in beginner EDA:
- Analyzing columns in isolation
- Comparing raw values instead of distributions
- Ignoring variability and only checking averages
- Drawing conclusions too early

Professional habit: put columns side by side and compare their statistical behavior.

```python
# Import library
import pandas as pd

print('Pandas version:', pd.__version__)
```

---

## Part 1: Load Data and Prepare Numeric Columns

```python
sales_df = pd.read_csv('../data/raw/sales_data.csv')

print('Sales data preview:')
print(sales_df.head())
print('\nShape:', sales_df.shape)
```

```python
numeric_cols = sales_df.select_dtypes(include='number').columns.tolist()
print('Numeric columns:', numeric_cols)
```

We will compare these three numeric columns: `OrderID`, `Quantity`, and `Price`.

---

## Part 2: Compare Central Tendency Across Columns

```python
central_tendency = sales_df[numeric_cols].agg(['count', 'mean', 'median']).T
print('Central tendency comparison:')
print(central_tendency)
```

```python
# Compare mean and median gap as a simple skew indicator
central_tendency['mean_minus_median'] = central_tendency['mean'] - central_tendency['median']
print('\nMean - Median gap:')
print(central_tendency[['mean', 'median', 'mean_minus_median']])
```

Interpretation prompt:
- Smaller mean-median gap suggests a more balanced distribution
- Larger gap can suggest skew or unusual values

---

## Part 3: Compare Spread and Variability Across Columns

```python
spread = sales_df[numeric_cols].agg(['min', 'max', 'std', 'var']).T
spread['range'] = spread['max'] - spread['min']

print('Spread comparison:')
print(spread[['min', 'max', 'range', 'std', 'var']])
```

```python
# Quick ranking by variability using standard deviation
spread_rank = spread['std'].sort_values(ascending=False)
print('Columns ranked by standard deviation (high to low):')
print(spread_rank)
```

Interpretation prompt:
- Higher std/var indicates higher variability
- Lower std/var indicates more stability

---

## Part 4: Identify Patterns and Possible Anomalies

```python
summary = sales_df[numeric_cols].describe().T
summary['range'] = summary['max'] - summary['min']
summary['mean_minus_median'] = summary['mean'] - summary['50%']

print('Distribution summary for anomaly scanning:')
print(summary[['count', 'mean', '50%', 'mean_minus_median', 'min', 'max', 'range', 'std']])
```

Questions to ask from this table:
- Which columns have unusually wide ranges?
- Which columns have noticeable mean-median gaps?
- Which variables seem stable versus volatile?
- What should be investigated next before conclusions?

---

## Part 5: Reusable Multi-Column Comparison Function

```python
def compare_distributions(df, columns):
    stats = df[columns].agg(['count', 'mean', 'median', 'min', 'max', 'std', 'var']).T
    stats['range'] = stats['max'] - stats['min']
    stats['mean_minus_median'] = stats['mean'] - stats['median']
    return stats

distribution_comparison = compare_distributions(sales_df, numeric_cols)
print(distribution_comparison[['count', 'mean', 'median', 'mean_minus_median', 'range', 'std', 'var']])
```
