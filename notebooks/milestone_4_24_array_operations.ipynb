## Part 2: Element-Wise Array Operations

When you perform operations on NumPy arrays, they are applied **element-by-element**.

This is called **element-wise computation**.

### Addition: Element-Wise

```python
# Create two arrays
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("Array 1:", arr1)
print("Array 2:", arr2)

# Add arrays element-wise
result = arr1 + arr2

print("\nArray 1 + Array 2:")
print(result)
print("\nExplanation:")
print(f"  {arr1[0]} + {arr2[0]} = {result[0]}")
print(f"  {arr1[1]} + {arr2[1]} = {result[1]}")
print(f"  {arr1[2]} + {arr2[2]} = {result[2]}")
print(f"  {arr1[3]} + {arr2[3]} = {result[3]}")

print("\n💡 Each element is added to its corresponding element!")
```

### Subtraction: Element-Wise

```python
arr1 = np.array([50, 40, 30, 20])
arr2 = np.array([5, 10, 15, 20])

print("Array 1:", arr1)
print("Array 2:", arr2)

# Subtract arrays element-wise
result = arr1 - arr2

print("\nArray 1 - Array 2:")
print(result)
print("\nBreakdown:")
print(f"  [50 - 5, 40 - 10, 30 - 15, 20 - 20]")
print(f"  [45, 30, 15, 0]")

print("\n💡 Subtraction also works element-wise!")
```

### Multiplication: Element-Wise

```python
arr1 = np.array([2, 3, 4, 5])
arr2 = np.array([10, 10, 10, 10])

print("Array 1:", arr1)
print("Array 2:", arr2)

# Multiply arrays element-wise
result = arr1 * arr2

print("\nArray 1 * Array 2:")
print(result)
print("\nCalculation:")
print(f"  [2*10, 3*10, 4*10, 5*10]")
print(f"  [20, 30, 40, 50]")

print("\n💡 NOT matrix multiplication - element-wise multiplication!")
```

### Division: Element-Wise

```python
arr1 = np.array([100, 50, 25, 10])
arr2 = np.array([10, 5, 5, 2])

print("Array 1:", arr1)
print("Array 2:", arr2)

# Divide arrays element-wise
result = arr1 / arr2

print("\nArray 1 / Array 2:")
print(result)
print("\nCalculation:")
print(f"  [100/10, 50/5, 25/5, 10/2]")
print(f"  [10.0, 10.0, 5.0, 5.0]")

print("\n💡 Division returns float results!")
```

### All Operations Together

```python
a = np.array([10, 20, 30, 40])
b = np.array([2, 4, 6, 8])

print("Array a:", a)
print("Array b:", b)

print("\n=== ELEMENT-WISE OPERATIONS ===")
print(f"\na + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a ** 2 = {a ** 2}  (power operation)")
print(f"a % 3 = {a % 3}  (modulo operation)")

print("\n💡 All operations work element-wise!")
```

### 2D Array Operations

```python
matrix1 = np.array([[1, 2, 3],
                    [4, 5, 6]])

matrix2 = np.array([[10, 10, 10],
                    [20, 20, 20]])

print("Matrix 1:")
print(matrix1)
print("\nMatrix 2:")
print(matrix2)

print("\n--- ADDITION ---")
print(matrix1 + matrix2)

print("\n--- MULTIPLICATION ---")
print(matrix1 * matrix2)

print("\n💡 Element-wise operations work on any shape!")
```

## Part 3: Scalar Operations on Arrays

You can apply a **single value** to an **entire array**.

This is called **scalar broadcasting**.

### Adding a Scalar

```python
arr = np.array([10, 20, 30, 40, 50])

print("Original array:", arr)

# Add 5 to every element
result = arr + 5

print("Array + 5:", result)
print("\nExplanation:")
print("  The scalar 5 is added to EVERY element")
print(f"  [10+5, 20+5, 30+5, 40+5, 50+5]")
print(f"  [15, 25, 35, 45, 55]")

print("\n💡 No loop needed - NumPy broadcasts the scalar!")
```

### Multiplying by a Scalar

```python
arr = np.array([1, 2, 3, 4, 5])

print("Original array:", arr)

# Multiply every element by 10
result = arr * 10

print("Array * 10:", result)
print("\nBreakdown:")
print("  [1*10, 2*10, 3*10, 4*10, 5*10]")
print("  [10, 20, 30, 40, 50]")

print("\n💡 Scalar multiplication applies to all elements!")
```

### All Scalar Operations

```python
arr = np.array([10, 20, 30, 40, 50])

print("Original array:", arr)

print("\n=== SCALAR OPERATIONS ===")
print(f"\narr + 100 = {arr + 100}")
print(f"arr - 5   = {arr - 5}")
print(f"arr * 2   = {arr * 2}")
print(f"arr / 10  = {arr / 10}")
print(f"arr ** 2  = {arr ** 2}")

print("\n💡 Every operation broadcasts the scalar to all elements!")
```

### Practical Example: Temperature Conversion

```python
# Temperatures in Celsius
celsius = np.array([0, 10, 20, 30, 40, 100])

print("Temperatures in Celsius:", celsius)

# Convert to Fahrenheit: F = C * 9/5 + 32
fahrenheit = celsius * 9/5 + 32

print("Temperatures in Fahrenheit:", fahrenheit)

print("\n--- COMPARISON ---")
for c, f in zip(celsius, fahrenheit):
    print(f"{c:3d}°C = {f:5.1f}°F")

print("\n💡 Formula applied to entire array in ONE line!")
```

### Practical Example: Price Discount

```python
# Original prices
prices = np.array([100, 200, 350, 500, 1000])

print("Original Prices:", prices)

# Apply 20% discount
discount = 0.20
discounted_prices = prices * (1 - discount)

print(f"Prices after {discount*100:.0f}% discount:", discounted_prices)

# Calculate savings
savings = prices - discounted_prices

print("Savings:", savings)

print("\n--- BREAKDOWN ---")
for orig, disc, save in zip(prices, discounted_prices, savings):
    print(f"${orig:4.0f} → ${disc:5.1f} (save ${save:4.1f})")

print("\n💡 Complex calculations with simple array operations!")
```

## Part 4: Comparing NumPy Arrays vs Python Lists

**Critical difference:** Lists and arrays behave differently with math operators!

### List Addition vs Array Addition

```python
# Python lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]

print("=== PYTHON LISTS ===")
print("list1:", list1)
print("list2:", list2)
print("\nlist1 + list2 =", list1 + list2)
print("Result: Lists are CONCATENATED (joined together)")

print("\n" + "="*50)

# NumPy arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("\n=== NUMPY ARRAYS ===")
print("arr1:", arr1)
print("arr2:", arr2)
print("\narr1 + arr2 =", arr1 + arr2)
print("Result: Elements are ADDED mathematically")

print("\n💡 HUGE DIFFERENCE: + means concatenation for lists, addition for arrays!")
```

### List Multiplication vs Array Multiplication

```python
# Python list
my_list = [1, 2, 3]

print("=== PYTHON LIST ===")
print("my_list:", my_list)
print("my_list * 3 =", my_list * 3)
print("Result: List is REPEATED 3 times")

print("\n" + "="*50)

# NumPy array
my_array = np.array([1, 2, 3])

print("\n=== NUMPY ARRAY ===")
print("my_array:", my_array)
print("my_array * 3 =", my_array * 3)
print("Result: Each element is MULTIPLIED by 3")

print("\n💡 Lists repeat, arrays multiply!")
```

### Why Lists Need Loops for Math

```python
# With Python lists - VERBOSE
list1 = [10, 20, 30, 40]
list2 = [1, 2, 3, 4]

print("=== PYTHON LISTS (manual loop needed) ===")
result_list = []
for i in range(len(list1)):
    result_list.append(list1[i] + list2[i])

print("list1:", list1)
print("list2:", list2)
print("Result:", result_list)
print("Code: 4 lines, loop required")

print("\n" + "="*50)

# With NumPy arrays - CONCISE
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("\n=== NUMPY ARRAYS (automatic element-wise) ===")
result_array = arr1 + arr2

print("arr1:", arr1)
print("arr2:", arr2)
print("Result:", result_array)
print("Code: 1 line, no loop")

print("\n💡 NumPy is cleaner, faster, and more expressive!")
```

## Part 5: Avoiding Common Mistakes

### Mistake 1: Shape Mismatch

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([10, 20, 30, 40])  # Different size!

print("Array 1 shape:", arr1.shape)
print("Array 2 shape:", arr2.shape)

print("\n❌ INCOMPATIBLE SHAPES")
print("arr1 + arr2 would cause an error:")
print("ValueError: operands could not be broadcast together")

print("\n✅ SOLUTION:")
print("Ensure arrays have the same shape for element-wise operations")
print("Or use broadcasting rules intentionally")

# Correct example
arr1_fixed = np.array([1, 2, 3, 4])
result = arr1_fixed + arr2
print(f"\nWith matching shapes: {arr1_fixed} + {arr2} = {result}")

print("\n💡 Always check shapes when working with multiple arrays!")
```

### Mistake 2: Assuming List Behavior

```python
print("❌ WRONG ASSUMPTION:")
print("'NumPy arrays work like lists'")
print("\nExample of confusion:")

my_list = [1, 2, 3]
my_array = np.array([1, 2, 3])

print(f"\nList * 2 = {my_list * 2}  (repetition)")
print(f"Array * 2 = {my_array * 2}  (multiplication)")

print("\n✅ CORRECT UNDERSTANDING:")
print("Lists are for collections, arrays are for numerical computation")
print("Operators have mathematical meaning for arrays")

print("\n💡 Don't expect list behavior from arrays!")
```

### Mistake 3: Using Loops Unnecessarily

```python
arr = np.array([1, 2, 3, 4, 5])

print("❌ INEFFICIENT (loop):")
result_bad = np.zeros(5)
for i in range(len(arr)):
    result_bad[i] = arr[i] * 2
print("Code: Loop with 3 lines")
print("Result:", result_bad)

print("\n✅ EFFICIENT (vectorized):")
result_good = arr * 2
print("Code: 1 line, no loop")
print("Result:", result_good)

print("\n💡 When you see a loop on arrays, ask: 'Can this be vectorized?'")
```

### Mistake 4: Ignoring Data Types

```python
# Integer division (Python 2 style)
arr_int = np.array([10, 20, 30], dtype=int)

print("Integer array:", arr_int)
print("Data type:", arr_int.dtype)

result = arr_int / 3
print("\narr_int / 3 =", result)
print("Result type:", result.dtype)
print("Note: Division returns floats by default")

# Integer-only division
result_int = arr_int // 3
print("\narr_int // 3 =", result_int)
print("Result type:", result_int.dtype)
print("Note: Floor division keeps integers")

print("\n💡 Be aware of data types in operations!")
```

## Part 6: Combining Operations

### Complex Formulas in One Line

```python
# Example: Calculate area of circles with different radii
radii = np.array([1, 2, 3, 4, 5])

print("Circle radii:", radii)

# Formula: Area = π * r²
areas = np.pi * radii ** 2

print("\nCircle areas:")
for r, a in zip(radii, areas):
    print(f"  Radius {r}: {a:.2f}")

print("\n💡 Complex formula applied to all values in ONE line!")
```

### Multiple Operations

```python
# Example: Normalize data to 0-1 range
data = np.array([10, 20, 30, 40, 50])

print("Original data:", data)
print(f"Min: {data.min()}, Max: {data.max()}")

# Normalize: (x - min) / (max - min)
normalized = (data - data.min()) / (data.max() - data.min())

print("\nNormalized data:", normalized)
print("Note: Values now range from 0 to 1")

print("\n--- VERIFICATION ---")
for orig, norm in zip(data, normalized):
    print(f"  {orig:2d} → {norm:.2f}")

print("\n💡 Common data science operation in a single formula!")
```

### Practical Example: Grade Scaling

```python
# Test scores out of 80
scores = np.array([72, 64, 68, 76, 80])

print("Original scores (out of 80):", scores)

# Convert to percentage
percentages = (scores / 80) * 100

print("Percentages:", percentages)

# Add 5 point curve
curved = percentages + 5

print("After 5-point curve:", curved)

# Cap at 100
final = np.minimum(curved, 100)

print("Final scores (capped at 100):", final)

print("\n--- SUMMARY ---")
for s, p, c, f in zip(scores, percentages, curved, final):
    print(f"  {s}/80 → {p:.1f}% → +5 → {f:.1f}%")

print("\n💡 Multiple transformations with array operations!")
```
