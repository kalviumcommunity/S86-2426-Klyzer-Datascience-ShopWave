# Milestone 4.31: Understanding Data Shapes and Column Data Types

**Objective:** Inspect DataFrame structure using `shape` and `dtypes` before any cleaning or analysis.

**Focus:**
- What shape means
- Rows vs columns
- Column data types
- Early type-related issue detection

```python
import pandas as pd

print('Pandas version:', pd.__version__)
```

```python
# Load a real dataset for structure inspection
df = pd.read_csv('../data/raw/sales_data.csv')

print('Loaded sales_data.csv successfully.')
```

## 1) Inspect Shape
`shape` returns `(rows, columns)`.

```python
print('Shape:', df.shape)
rows, cols = df.shape
print('Rows (observations):', rows)
print('Columns (features):', cols)
```

## 2) Inspect Column Data Types
`dtypes` shows the data type for every column.

```python
print(df.dtypes)

print('\nType counts:')
print(df.dtypes.value_counts())
```

## 3) Understand Rows and Columns Clearly

```python
print('Columns list:')
print(df.columns.tolist())

print('\nFirst 3 records:')
print(df.head(3))
```

## 4) Detect Type-Related Risk Signals Early
No cleaning is done here. This is only awareness and inspection.

```python
print('Missing values per column:')
print(df.isna().sum())

text_like_cols = df.select_dtypes(include=['object', 'string', 'str']).columns.tolist()
print('\nText-like columns (object/string/str):', text_like_cols)

print('\nQuick interpretation:')
print('- Numeric dtypes support arithmetic directly.')
print('- Text-like dtypes often contain labels, categories, or mixed values and should be reviewed carefully.')
print('- Missing values can affect behavior of downstream operations.')
```

```python
# Optional advanced check: find text-like columns that look mostly numeric
# This does not transform data; it only flags suspicious columns.
suspicious = []
for col in text_like_cols:
    numeric_ratio = pd.to_numeric(df[col], errors='coerce').notna().mean()
    if numeric_ratio >= 0.8:
        suspicious.append((col, numeric_ratio))

print('Suspicious text-like columns (>=80% numeric-like):', suspicious if suspicious else 'None found')
```
