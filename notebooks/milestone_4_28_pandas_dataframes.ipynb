## Part 1: Understanding Pandas DataFrames

A **DataFrame** is Pandas' primary data structure for working with tabular (table-like) data.

**Key Features**:
- **2-dimensional** (rows and columns)
- **Columns** have names (like spreadsheet headers)
- **Rows** have indexes (like row numbers)
- **Each column** is a Series
- **Mixed data types** across columns

**Real-World Analogy**:
- Spreadsheet (Excel/Google Sheets)
- Database table (SQL)
- CSV file when opened

**Why DataFrames Matter**:
- 90% of data analysis uses DataFrames
- Natural representation of tabular data
- Powerful operations on rows and columns
- Integration with files, databases, APIs

### Your First DataFrame

````python
# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

print("My First DataFrame:")
print(df)
print(f"\nType: {type(df)}")
````

### Understanding the Structure

When you print a DataFrame, you see:

```
     Name  Age         City      ← Column names (headers)
0   Alice   25     New York      ← Row 0 (index 0)
1     Bob   30  Los Angeles      ← Row 1 (index 1)
2 Charlie   22      Chicago      ← Row 2 (index 2)
```

**Components**:
- **Index** (left): Row labels (0, 1, 2...)
- **Columns** (top): Column names
- **Values** (center): Actual data

````python
# Inspect DataFrame components
print("Column names:")
print(df.columns)

print("\nIndex:")
print(df.index)

print("\nValues (as NumPy array):")
print(df.values)

print("\nShape (rows, columns):")
print(df.shape)
````

### DataFrame vs Series

**Series** = 1D labeled data (single column)
**DataFrame** = 2D labeled data (multiple columns)

A DataFrame is essentially multiple Series sharing the same index!

````python
# A DataFrame is made of Series
print("'Name' column (Series):")
print(df['Name'])
print(f"Type: {type(df['Name'])}")

print("\n'Age' column (Series):")
print(df['Age'])
print(f"Type: {type(df['Age'])}")
````

## Part 2: Creating DataFrames from Dictionaries

The most common way to create a DataFrame programmatically is from a Python dictionary.

### Method 1: Dictionary with Lists (Column-Oriented)

````python
# Dictionary where keys are column names, values are lists
student_data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Emma'],
    'Math': [85, 92, 78, 95, 88],
    'English': [90, 88, 85, 92, 94],
    'Science': [88, 90, 82, 96, 91]
}

# Create DataFrame
students_df = pd.DataFrame(student_data)

print("Student Grades DataFrame:")
print(students_df)
````

### How It Works

```python
{
    'Column1': [value1, value2, value3],  ← Column 1
    'Column2': [value1, value2, value3],  ← Column 2
    'Column3': [value1, value2, value3]   ← Column 3
}
```

- **Keys** → Column names
- **Values** (lists) → Column data
- All lists must have **same length**

````python
# Inspect the resulting DataFrame
print("Shape:", students_df.shape)
print("\nColumn names:", list(students_df.columns))
print("\nData types:")
print(students_df.dtypes)
````

### Method 2: List of Dictionaries (Row-Oriented)

````python
# List where each element is a dictionary representing a row
employees = [
    {'Name': 'John', 'Department': 'Sales', 'Salary': 50000},
    {'Name': 'Sarah', 'Department': 'IT', 'Salary': 65000},
    {'Name': 'Mike', 'Department': 'HR', 'Salary': 55000},
    {'Name': 'Lisa', 'Department': 'IT', 'Salary': 70000}
]

# Create DataFrame
employees_df = pd.DataFrame(employees)

print("Employee DataFrame:")
print(employees_df)
````

### How It Works (Row-Oriented)

```python
[
    {'col1': val1, 'col2': val2},  ← Row 1
    {'col1': val3, 'col2': val4},  ← Row 2
    {'col1': val5, 'col2': val6}   ← Row 3
]
```

- Each dictionary is a **row**
- Dictionary keys become **column names**
- Useful for JSON-like data

### Example: Product Catalog

````python
# Create a product catalog
products = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Price': [999.99, 25.50, 75.00, 299.99, 150.00],
    'Stock': [15, 120, 85, 30, 60],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Accessories']
}

products_df = pd.DataFrame(products)

print("Product Catalog:")
print(products_df)

print("\nDataFrame Info:")
print(f"Rows: {len(products_df)}")
print(f"Columns: {len(products_df.columns)}")
print(f"Total cells: {products_df.size}")
````

### Custom Index

````python
# Create DataFrame with custom index
temps = {
    'Morning': [20, 18, 22, 19, 21],
    'Afternoon': [28, 26, 30, 27, 29],
    'Evening': [24, 22, 26, 23, 25]
}

temps_df = pd.DataFrame(temps, index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])

print("Temperature DataFrame (Custom Index):")
print(temps_df)
````

## Part 3: Loading DataFrames from Files

In real-world scenarios, data usually comes from external files. CSV is the most common format.

### Loading a CSV File

````python
# Load CSV file
# Note: Adjust the path to where your file is located
csv_path = '../data/raw/sample_data.csv'

df_from_csv = pd.read_csv(csv_path)

print("DataFrame loaded from CSV:")
print(df_from_csv)
print(f"\nShape: {df_from_csv.shape}")
````

### How CSV Loading Works

**CSV File Content:**
```
Name,Age,City,Score      ← First row = column names (header)
Alice,25,New York,85     ← Data rows
Bob,30,Los Angeles,92
Charlie,22,Chicago,78
```

**`pd.read_csv()` automatically:**
- Reads the first row as column names
- Creates numeric index (0, 1, 2...)
- Infers data types for each column
- Handles common delimiters (comma by default)

### Creating a CSV for Practice

````python
# Create a sample DataFrame
practice_data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Sales': [1200, 1500, 1350, 1800, 1650],
    'Customers': [45, 60, 52, 75, 68],
    'Returns': [2, 3, 1, 4, 2]
}

practice_df = pd.DataFrame(practice_data)

# Save to CSV
output_path = '../data/processed/practice_sales.csv'
practice_df.to_csv(output_path, index=False)

print("CSV file created successfully!")
print(f"Saved to: {output_path}")
print("\nOriginal DataFrame:")
print(practice_df)
````

````python
# Load it back to verify
loaded_df = pd.read_csv(output_path)

print("Loaded DataFrame:")
print(loaded_df)

print("\n✅ Round-trip successful: Created CSV and loaded it back!")
````

## Part 4: Inspecting DataFrame Structure

After loading data, **always inspect it** before analysis.

### Method 1: `.head()` and `.tail()`

````python
# Create a larger DataFrame
large_data = {
    'ID': range(1, 21),
    'Value': np.random.randint(50, 100, 20),
    'Category': np.random.choice(['A', 'B', 'C'], 20)
}

large_df = pd.DataFrame(large_data)

print("First 5 rows:")
print(large_df.head())

print("\nLast 5 rows:")
print(large_df.tail())

print("\nFirst 3 rows:")
print(large_df.head(3))
````

### Method 2: `.info()`

````python
# Get comprehensive information
print("DataFrame Info:")
large_df.info()

print("\nWhat .info() shows:")
print("- Number of rows and columns")
print("- Column names")
print("- Non-null counts (missing data detection)")
print("- Data types")
print("- Memory usage")
````

### Method 3: `.describe()`

````python
# Statistical summary for numeric columns
print("Statistical Summary:")
print(large_df.describe())

print("\nWhat .describe() shows:")
print("- count: Number of non-null values")
print("- mean: Average")
print("- std: Standard deviation")
print("- min/max: Range")
print("- 25%, 50%, 75%: Quartiles")
````

### Method 4: Shape, Columns, and Index

````python
# Basic structure inspection
print("Shape (rows, columns):", large_df.shape)
print(f"Number of rows: {large_df.shape[0]}")
print(f"Number of columns: {large_df.shape[1]}")

print("\nColumn names:")
print(list(large_df.columns))

print("\nIndex:")
print(large_df.index)

print("\nData types:")
print(large_df.dtypes)
````

### Method 5: Checking for Missing Data

````python
# Create DataFrame with missing values
data_with_nulls = {
    'A': [1, 2, None, 4, 5],
    'B': [10, None, 30, 40, 50],
    'C': [100, 200, 300, None, 500]
}

df_nulls = pd.DataFrame(data_with_nulls)

print("DataFrame with missing values:")
print(df_nulls)

print("\nCheck for null values:")
print(df_nulls.isnull())

print("\nCount of null values per column:")
print(df_nulls.isnull().sum())

print("\nTotal null values:")
print(df_nulls.isnull().sum().sum())
````

## Part 5: Accessing Data in DataFrames

Learn various ways to access and select data.

### Selecting Columns

````python
# Use the student grades DataFrame
print("Original DataFrame:")
print(students_df)

# Select single column (returns Series)
print("\n'Math' column (Series):")
print(students_df['Math'])
print(f"Type: {type(students_df['Math'])}")

# Select multiple columns (returns DataFrame)
print("\n'Student' and 'Math' columns (DataFrame):")
print(students_df[['Student', 'Math']])
print(f"Type: {type(students_df[['Student', 'Math']])}")
````

### Selecting Rows

````python
# By position (iloc)
print("First row (iloc):")
print(students_df.iloc[0])

print("\nFirst two rows:")
print(students_df.iloc[0:2])

# By label (loc)
print("\nRow with index 0 (loc):")
print(students_df.loc[0])
````

### Selecting Specific Cells

````python
# Access specific cell
print("Cell at row 0, column 'Math':")
print(students_df.loc[0, 'Math'])

# Access by position
print("\nCell at row 1, column 2 (by position):")
print(students_df.iloc[1, 2])

# Access multiple cells
print("\nFirst 3 rows, columns 'Student' and 'Math':")
print(students_df.loc[0:2, ['Student', 'Math']])
````

### Filtering Rows (Boolean Indexing)

````python
# Filter students with Math score > 85
print("Students with Math > 85:")
high_math = students_df[students_df['Math'] > 85]
print(high_math)

# Multiple conditions
print("\nStudents with Math > 85 AND English > 90:")
top_students = students_df[(students_df['Math'] > 85) & (students_df['English'] > 90)]
print(top_students)
````

## Part 6: Real-World Examples

Practical scenarios for DataFrame creation and inspection.

### Example 1: Monthly Sales Report

````python
# Create monthly sales data
sales_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [45000, 52000, 48000, 61000, 58000, 64000],
    'Expenses': [32000, 35000, 33000, 41000, 39000, 42000],
    'Customers': [450, 520, 480, 610, 580, 640]
}

sales_df = pd.DataFrame(sales_data)

print("Monthly Sales Report:")
print(sales_df)

# Calculate profit
sales_df['Profit'] = sales_df['Revenue'] - sales_df['Expenses']

print("\nWith Profit Column:")
print(sales_df)

# Summary statistics
print("\nSummary:")
print(f"Total Revenue: ${sales_df['Revenue'].sum():,}")
print(f"Average Profit: ${sales_df['Profit'].mean():,.2f}")
print(f"Best Month: {sales_df.loc[sales_df['Profit'].idxmax(), 'Month']}")
````

### Example 2: Student Performance Analysis

````python
# More detailed student data
student_records = {
    'StudentID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'Diana Ross', 'Emma Watson'],
    'Math': [92, 85, 78, 95, 88],
    'English': [88, 90, 85, 92, 94],
    'Science': [90, 87, 82, 96, 91],
    'Grade': ['A', 'B', 'C', 'A', 'B']
}

students = pd.DataFrame(student_records)

print("Student Records:")
print(students)

# Calculate average score
students['Average'] = students[['Math', 'English', 'Science']].mean(axis=1)

print("\nWith Average Scores:")
print(students)

# Find top performer
top_student = students.loc[students['Average'].idxmax()]
print(f"\nTop Student: {top_student['Name']} (Average: {top_student['Average']:.2f})")
````

### Example 3: Inventory Management

````python
# Inventory data
inventory = {
    'ProductID': ['P001', 'P002', 'P003', 'P004', 'P005'],
    'ProductName': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Price': [999.99, 25.50, 75.00, 299.99, 150.00],
    'Stock': [15, 120, 85, 30, 60],
    'ReorderLevel': [10, 50, 40, 20, 30]
}

inventory_df = pd.DataFrame(inventory)

print("Inventory:")
print(inventory_df)

# Calculate total value
inventory_df['TotalValue'] = inventory_df['Price'] * inventory_df['Stock']

print("\nWith Total Value:")
print(inventory_df)

# Check reorder needs
print("\nProducts Needing Reorder:")
reorder_needed = inventory_df[inventory_df['Stock'] < inventory_df['ReorderLevel']]
if len(reorder_needed) > 0:
    print(reorder_needed[['ProductName', 'Stock', 'ReorderLevel']])
else:
    print("All products are well-stocked!")

print(f"\nTotal Inventory Value: ${inventory_df['TotalValue'].sum():,.2f}")
````

## Summary and Best Practices

### Key Takeaways

1. **DataFrame = 2D Labeled Data** - Rows and columns with labels
2. **From Dictionary** - `pd.DataFrame(dict)` for programmatic creation
3. **From CSV** - `pd.read_csv()` for file loading
4. **Always Inspect** - Use `.head()`, `.info()`, `.describe()`
5. **Columns are Series** - Each column is a separate Series
6. **Mixed Types OK** - Different columns can have different dtypes

### Best Practices

```python
# ✓ DO: Always inspect after loading
df = pd.read_csv('data.csv')
print(df.head())
df.info()

# ✓ DO: Check for missing data
print(df.isnull().sum())

# ✓ DO: Verify shape and columns
print(df.shape)
print(df.columns)

# ✗ DON'T: Assume column names are correct
# Always check: print(df.columns)

# ✗ DON'T: Forget to specify file path correctly
# Use relative or absolute paths

# ✗ DON'T: Skip inspection
# Silent errors are the worst!
```

### Essential DataFrame Methods

| Method | Purpose | Example |
|--------|---------|--------|
| `.head(n)` | First n rows | `df.head(5)` |
| `.tail(n)` | Last n rows | `df.tail(3)` |
| `.info()` | Column info, types, nulls | `df.info()` |
| `.describe()` | Statistical summary | `df.describe()` |
| `.shape` | (rows, columns) | `df.shape` |
| `.columns` | Column names | `df.columns` |
| `.dtypes` | Data types | `df.dtypes` |
| `.isnull()` | Check for missing | `df.isnull().sum()` |

### When to Use DataFrames

✅ **Use DataFrames when:**
- Working with tabular data
- Loading CSV, Excel, or database data
- Need to analyze multiple columns together
- Performing data cleaning and transformation
- Creating reports or visualizations

❌ **Don't use DataFrames when:**
- Only need a single column (use Series)
- Working with pure numerical computation (use NumPy)
- Data doesn't have tabular structure

### The DataFrame Mindset

Think:
> "My data is a table with meaningful rows and columns"

Not:
> "I need a collection of lists"

**DataFrames are the working table for all data analysis - master them and you master Pandas!**
