**Objective:** Master the three essential DataFrame inspection methods that every Data Scientist uses before analyzing data.

**Key Learning:**
- Inspection is **mandatory** after loading data

## Why DataFrame Inspection Matters

### The Problem

**Common beginner mistakes:**
- ❌ Loading data and immediately starting analysis
- ❌ Assuming column types are correct
- ❌ Missing null values that break operations
- ❌ Not checking data ranges before calculations
- ❌ Trusting data without verification

**Result:** Hours wasted debugging errors that inspection would catch in seconds.

### The Solution

**Three inspection methods:**

1. **`head()`** → "What does my data look like?"
   - Quick visual preview
   - See actual values
   - Check column names

2. **`info()`** → "What is my data structure?"
   - Data types
   - Null counts
   - Memory usage

3. **`describe()`** → "What are my data distributions?"
   - Statistics (mean, std, min, max)
   - Percentiles
   - Outlier hints

**Professional habit:** Use all three, every time.

## Part 1: Inspecting Data with head()

### What is head()?

**Purpose:** Preview the **first few rows** of a DataFrame

**Why use it:**
- Quick visual confirmation data loaded correctly
- See actual values, not just statistics
- Check column names are as expected
- Verify data alignment (values in correct columns)

**Think of it like:** Glancing at the first page of a book to see if it's the right book.

### Loading Sample Data

````python
# Load employee dataset
employees_df = pd.read_csv('../data/raw/employees.csv')

print("✓ Employees data loaded")
print(f"Shape: {employees_df.shape}")
````

### Using head() - Default Behavior

````python
# Default: first 5 rows
print("First 5 rows (default):")
print(employees_df.head())
````

**What to observe:**
- ✅ Column names at the top
- ✅ Index on the left (0, 1, 2, 3, 4)
- ✅ Actual data values
- ✅ Data types (numeric vs text)
- ✅ Any obvious issues (duplicates, NaN, weird values)

### Using head() with Custom Row Count

````python
# First 3 rows
print("First 3 rows:")
print(employees_df.head(3))
````

````python
# First 10 rows
print("First 10 rows:")
print(employees_df.head(10))
````

**When to use different row counts:**
- **`head(3)`** → Quick glance, tight on space
- **`head()`** → Standard preview (5 rows)
- **`head(10)`** → More comprehensive view
- **`head(20)`** → Looking for patterns across many rows

### Comparing head() with tail()

````python
# First 3 rows
print("FIRST 3 rows:")
print(employees_df.head(3))

# Last 3 rows
print("\nLAST 3 rows:")
print(employees_df.tail(3))
````

**Why check both ends:**
- Patterns might differ at start vs end
- File loading issues often show at the end
- Verify entire file loaded (not truncated)
- See full range of data

### What head() Reveals

**From the employees data, we can see:**

1. **Column Names:**
   - EmployeeID, Name, Department, Salary, YearsExperience, City

2. **Data Types (visual clues):**
   - EmployeeID: Numbers (likely int)
   - Name: Text (object/string)
   - Department: Text (object/string)
   - Salary: Numbers (likely int)
   - YearsExperience: Numbers (likely int)
   - City: Text (object/string)

3. **Sample Values:**
   - Salaries range from 62,000 to 110,000
   - Multiple departments: Engineering, Marketing, Sales, HR
   - Cities are diverse: New York, Los Angeles, etc.

4. **No obvious issues:**
   - No NaN visible (but check info() to be sure)
   - No weird characters or formatting
   - Data aligned correctly

## Part 2: Inspecting Structure with info()

### What is info()?

**Purpose:** Get a **comprehensive overview** of DataFrame structure

**What it shows:**
- Number of rows and columns
- Column names
- Data types for each column
- Non-null counts (detects missing data)
- Memory usage

**Think of it like:** Reading a car's diagnostic report—tells you everything about the system's health.

**This is the MOST IMPORTANT inspection method!**

### Using info() on Employees Data

````python
# Get comprehensive information
print("DataFrame Information:")
employees_df.info()
````

### Reading info() Output

**Let's break down what info() tells us:**

```
<class 'pandas.core.frame.DataFrame'>
RangeIndex: 10 entries, 0 to 9              ← 10 rows total
Data columns (total 6 columns):             ← 6 columns total
 #   Column            Non-Null Count  Dtype   
---  ------            --------------  -----   
 0   EmployeeID        10 non-null     int64   ← All 10 rows have values (no nulls)
 1   Name              10 non-null     object  ← Text data
 2   Department        10 non-null     object  ← Text data
 3   Salary            10 non-null     int64   ← Numeric (integer)
 4   YearsExperience   10 non-null     int64   ← Numeric (integer)
 5   City              10 non-null     object  ← Text data
dtypes: int64(3), object(3)                   ← 3 numeric, 3 text columns

```

**Key insights:**
1. **10 rows, 6 columns** - Shape confirmed
2. **All columns have 10 non-null** - No missing data!
3. **Data types are correct:**
   - IDs, Salary, Experience → int64 (good)

### Detecting Missing Data with info()

````python
# Load sales data (has missing values)
sales_df = pd.read_csv('../data/raw/sales_data.csv')

print("Sales DataFrame Information:")
sales_df.info()
````

**Critical observation:**


**This means:**
- 2 rows have missing city information

### Understanding Data Types

**Common Pandas data types:**

| dtype | Meaning | Example |
|-------|---------|----------|
| **int64** | Integer numbers | 1, 42, 1000 |
| **float64** | Decimal numbers | 3.14, 99.99, 1.5 |
| **object** | Text/strings | "Alice", "Sales", "New York" |
| **bool** | True/False | True, False |
| **datetime64** | Dates/times | 2024-01-15 |
| **category** | Categorical data | "Low", "Medium", "High" |

**Why data types matter:**
- **int64/float64** → Can do math operations (+, -, *, /)
- **object** → Text operations (upper, lower, split)
- **datetime64** → Date operations (month, year, diff)

**Common issue:**
- Numbers stored as 'object' → Can't do calculations
- Dates as 'object' → Can't sort chronologically
- **Solution:** Convert types explicitly if needed

````python
# Check data types directly
print("Employees data types:")
print(employees_df.dtypes)

print("\nSales data types:")
print(sales_df.dtypes)
````

## Part 3: Summarizing Data with describe()

### What is describe()?

**Purpose:** Get **statistical summary** of numeric columns

**What it shows:**
- **count** - Number of non-null values
- **mean** - Average value
- **std** - Standard deviation (spread)
- **min** - Minimum value
- **25%** - First quartile (25th percentile)
- **50%** - Median (50th percentile)
- **75%** - Third quartile (75th percentile)
- **max** - Maximum value

**Think of it like:** Getting a health checkup's vital signs—tells you if values are in healthy range.

**Note:** Only works on numeric columns by default!

### Using describe() on Employees Data

````python
# Get statistical summary
print("Statistical Summary of Employees:")
print(employees_df.describe())
````

### Interpreting describe() Output

**Let's analyze the Salary column:**

```
           Salary
count       10.0        ← 10 employees (all have salaries)
mean     80800.0        ← Average salary: $80,800
std      15947.5        ← Salaries vary by ~$16,000
min      62000.0        ← Lowest salary: $62,000
25%      68500.0        ← 25% earn $68,500 or less
50%      75000.0        ← Median salary: $75,000 (middle value)
75%      91500.0        ← 75% earn $91,500 or less
max     110000.0        ← Highest salary: $110,000
```

**Key insights:**
1. **Range:** $62K to $110K (spread of $48K)
2. **Average:** $80.8K (mean)
3. **Typical value:** $75K (median, 50%)
4. **Spread:** Standard deviation of $16K suggests moderate variation
5. **No obvious outliers** (max isn't extreme compared to quartiles)

### Understanding Percentiles

**What do percentiles mean?**

- **25th percentile (Q1):** 25% of values are below this, 75% above
- **50th percentile (Median):** Middle value—half below, half above
- **75th percentile (Q3):** 75% of values are below this, 25% above

**Example with Salary:**
- 25% = $68,500 → 25% of employees earn $68,500 or less
- 50% = $75,000 → Half earn less than $75K, half earn more
- 75% = $91,500 → 75% earn $91,500 or less

**Why percentiles matter:**
- More robust than mean (not affected by outliers)
- Show distribution shape
- Help spot skewed data

````python
# View specific statistics
print("Mean Salary:", employees_df['Salary'].mean())
print("Median Salary:", employees_df['Salary'].median())
print("Min Salary:", employees_df['Salary'].min())
print("Max Salary:", employees_df['Salary'].max())
print("Std Dev:", employees_df['Salary'].std())
````

### Detecting Outliers with describe()

````python
# Examine YearsExperience
print("Years of Experience Statistics:")
print(employees_df['YearsExperience'].describe())
````

**Analyzing YearsExperience:**

- **Range:** 2 to 8 years
- **Mean:** ~4.2 years
- **Median:** 4 years
- **No extreme outliers** (max of 8 is reasonable)

**How to spot outliers:**
- Compare min/max to quartiles
- If max >> 75th percentile → potential outlier on high end
- If min << 25th percentile → potential outlier on low end
- Large std relative to mean → high variability

### describe() on All Columns (Including Text)

````python
# Include all columns (not just numeric)
print("All columns summary:")
print(employees_df.describe(include='all'))
````

**For text columns (object dtype):**
- **count** - Non-null values
- **unique** - Number of distinct values
- **top** - Most frequent value
- **freq** - Frequency of most common value

**Example:**
```
Department:
  count: 10
  unique: 4           ← 4 different departments
  top: Engineering    ← Most common department
  freq: 4             ← 4 employees in Engineering
```

## Part 4: Knowing When to Use Each Method

### The Three-Method Inspection Workflow

**Best practice: Use all three methods in order!**

```python
# 1. Quick visual check
df.head()

# 2. Structural health check
df.info()

# 3. Statistical understanding
df.describe()
```

### When to Use head()

**Use head() when you need to:**
- ✅ Quickly verify data loaded correctly
- ✅ See actual values (not just statistics)
- ✅ Check column names visually
- ✅ Spot obvious formatting issues
- ✅ Verify data alignment
- ✅ Get a feel for the dataset

**What head() tells you:**
- What the data **looks like**
- Sample values for each column
- Visual confirmation of structure

**What head() DOESN'T tell you:**
- Exact data types
- Missing values counts
- Statistical distributions
- Total row count

### When to Use info()

**Use info() when you need to:**
- ✅ Understand data types for each column
- ✅ Detect missing values
- ✅ Verify column names
- ✅ Check total row count
- ✅ Assess memory usage
- ✅ Plan data cleaning steps

**What info() tells you:**
- **Structure** of the DataFrame
- **Health** of the data (nulls)
- **Types** of each column

**What info() DOESN'T tell you:**
- Actual values
- Statistical distributions
- Outliers or ranges

**💡 Pro tip:** info() is the MOST important inspection method!

### When to Use describe()

**Use describe() when you need to:**
- ✅ Understand numeric distributions
- ✅ Check for outliers
- ✅ See data ranges (min/max)
- ✅ Get quick statistics
- ✅ Compare columns numerically
- ✅ Validate expected ranges

**What describe() tells you:**
- **Distribution** of numeric values
- **Central tendency** (mean, median)
- **Spread** (std, quartiles)
- **Range** (min, max)

**What describe() DOESN'T tell you:**
- Data types
- Missing values location
- Text column content
- Actual values

### Comparison Table

| Method | Question Answered | Shows | Best For |
|--------|-------------------|-------|----------|
| **head()** | What does my data look like? | Sample rows | Quick visual check |
| **info()** | What is my data structure? | Types, nulls, memory | Detecting issues |
| **describe()** | What are my distributions? | Statistics, ranges | Understanding numbers |

**Use all three for complete understanding!**

### Complete Inspection Example

````python
# Complete inspection workflow
print("=" * 70)
print("COMPLETE DATAFRAME INSPECTION WORKFLOW")
print("=" * 70)

# Step 1: Quick preview
print("\n1. QUICK PREVIEW (head()):")
print("-" * 70)
print(employees_df.head(3))

# Step 2: Structural check
print("\n2. STRUCTURAL CHECK (info()):")
print("-" * 70)
employees_df.info()

# Step 3: Statistical summary
print("\n3. STATISTICAL SUMMARY (describe()):")
print("-" * 70)
print(employees_df.describe())

print("\n" + "=" * 70)
print("✓ INSPECTION COMPLETE - Ready to analyze!")
print("=" * 70)
````

## Part 5: Building Inspection Habits

### The Professional Inspection Routine

**Every time you load data, follow this routine:**

```python
# 1. Load data
df = pd.read_csv('file.csv')

# 2. IMMEDIATE inspection (MANDATORY!)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nPreview:")
print(df.head())
print("\nStructure:")
df.info()
print("\nStatistics:")
print(df.describe())

# 3. Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# 4. Only NOW proceed with analysis
```

**This takes 30 seconds but saves hours of debugging!**

### Common Inspection Mistakes to Avoid

**❌ Mistake 1: Skipping inspection**
```python
df = pd.read_csv('data.csv')
# Immediately start analysis without inspection
result = df['column'].mean()  # Could fail if nulls exist!
```

**✅ Correct:**
```python
df = pd.read_csv('data.csv')
df.info()  # Check structure first
df.describe()  # Understand distribution
# Now safe to analyze
result = df['column'].mean()
```

---

**❌ Mistake 2: Only using head()**
```python
df.head()  # Looks good!
# But might have nulls in row 100, or wrong types
```

**✅ Correct:**
```python
df.head()      # Visual check
df.info()      # Full structure
df.describe()  # Statistics
```

---

**❌ Mistake 3: Ignoring data types**
```python
# Not checking if 'Age' is numeric
df['Age'].mean()  # Fails if Age is 'object' type
```

**✅ Correct:**
```python
df.info()  # Shows Age is 'object'
df['Age'] = pd.to_numeric(df['Age'])  # Convert first
df['Age'].mean()  # Now works
```

---

**❌ Mistake 4: Not checking for missing values**
```python
df['column'].mean()  # Returns NaN if any nulls exist
```

**✅ Correct:**
```python
df.info()  # Shows non-null counts
df.isnull().sum()  # Explicit null check
# Then decide: drop nulls or fill them
```

### Real-World Inspection Example

````python
# Inspect sales data (with missing values)
print("=" * 70)
print("INSPECTING SALES DATA (Real-world scenario)")
print("=" * 70)

print("\n1. Preview (head):")
print(sales_df.head())

print("\n2. Structure (info):")
sales_df.info()

print("\n3. Statistics (describe):")
print(sales_df.describe())

print("\n4. Missing values check:")
print(sales_df.isnull().sum())

print("\n" + "=" * 70)
print("FINDINGS:")
print("  - 10 orders total")
print("  - CustomerCity has 2 missing values (rows 1 and 7)")
print("  - OrderDate is 'object' type (should convert to datetime)")
print("  - Quantity ranges from 1 to 10")
print("  - Price ranges from $149.95 to $1999.98")
print("\nNEXT STEPS:")
print("  1. Decide how to handle missing cities (drop or fill)")
print("  2. Convert OrderDate to datetime type")
print("  3. Proceed with analysis")
print("=" * 70)
````
