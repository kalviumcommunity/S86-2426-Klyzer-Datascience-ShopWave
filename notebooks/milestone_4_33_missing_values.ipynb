# Milestone 4.33: Detecting Missing Values in DataFrames

**Objective:** Detect and summarize missing values clearly before any cleaning or analysis.

**Focus:**
- Understand what missing values represent
- Detect missing values at DataFrame level
- Count missing values per column
- Inspect rows containing missing entries

```python
import pandas as pd

print('Pandas version:', pd.__version__)
```

```python
# Load dataset used for missing-value detection
df = pd.read_csv('../data/raw/sales_data.csv')

print('Loaded sales_data.csv successfully')
print('Shape:', df.shape)
print(df.head(3))
```

## 1) Understanding Missing Values
In Pandas, missing values are typically represented as `NaN`/`NA` depending on dtype.
A blank field in a CSV is often parsed as missing data.

```python
print('Data types:')
print(df.dtypes)

print('\nPreview to spot blank/missing fields:')
print(df[['OrderID', 'CustomerCity']])
```

## 2) Detecting Missing Values
Use `isna()` (or `isnull()`) to create a boolean mask where `True` means missing.

```python
missing_mask = df.isna()
print('Missing-value mask (first 5 rows):')
print(missing_mask.head())
```

```python
print('Columns that contain at least one missing value:')
print(df.columns[df.isna().any()].tolist())
```

## 3) Counting Missing Values
Summaries help judge severity before deciding any cleaning strategy.

```python
missing_count = df.isna().sum()
missing_percent = (missing_count / len(df) * 100).round(2)

missing_summary = pd.DataFrame({
    'missing_count': missing_count,
    'missing_percent': missing_percent
})

print('Missing-value summary by column:')
print(missing_summary)
```

## 4) Inspecting Rows with Missing Data
Row-level inspection gives context and helps avoid assumptions.

```python
rows_with_missing = df[df.isna().any(axis=1)]

print('Rows containing at least one missing value:')
print(rows_with_missing)
print('\nCount of rows with missing values:', len(rows_with_missing))
```

```python
print('Row-level missing count:')
print(df.isna().sum(axis=1))
```

## Milestone Summary
- You detected missing values using `isna()`.
- You identified columns with missing entries.
- You counted missing values and percentages by column.
- You inspected rows containing missing data.

This milestone is about **detection and interpretation**, not filling or dropping values.
