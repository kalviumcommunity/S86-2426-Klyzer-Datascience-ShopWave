# Milestone 4.32: Selecting Rows and Columns Using Indexing and Slicing

**Objective:** Practice safe and intentional selection of rows and columns in Pandas DataFrames.

**Focus:**
- Select columns by name
- Select rows by position (`iloc`)
- Select rows by label (`loc`)
- Combine row and column selection

```python
import pandas as pd

print('Pandas version:', pd.__version__)
```

```python
# Load dataset used for selection examples
df = pd.read_csv('../data/raw/employees.csv')

print('Loaded employees.csv successfully')
print('Shape:', df.shape)
```

## 1) Select Columns by Name

```python
# Single column -> Series
name_series = df['Name']
print('Type of df[Name]:', type(name_series))
print(name_series.head(3))
```

```python
# Multiple columns -> DataFrame
employee_summary = df[['Name', 'Department', 'Salary']]
print('Type of df[[Name, Department, Salary]]:', type(employee_summary))
print(employee_summary.head(3))
```

## 2) Select Rows by Position with `iloc`
`iloc` uses zero-based integer positions and stop-exclusive slicing.

```python
print('First row (position 0):')
print(df.iloc[0])
```

```python
print('Rows at positions 2 to 5 (2:6):')
print(df.iloc[2:6])

print('Rows at positions [1, 4, 7]:')
print(df.iloc[[1, 4, 7]])
```

## 3) Select Rows by Label with `loc`
Set `EmployeeID` as the index so labels are meaningful for row selection.

```python
df_label = df.set_index('EmployeeID')
print('New index name:', df_label.index.name)
print('Index labels:', df_label.index.tolist())
```

```python
print('Row with label 104:')
print(df_label.loc[104])

print('Rows with labels 104 through 107 (inclusive):')
print(df_label.loc[104:107])
```

## 4) Select Rows and Columns Together
Use one clear expression to avoid ambiguous chained indexing.

```python
print('Position-based subset: rows 1:5, columns [1,2,3]')
print(df.iloc[1:5, [1, 2, 3]])
```

```python
print('Label-based subset: EmployeeID 103:106, columns Name and Department')
subset = df_label.loc[103:106, ['Name', 'Department']]
print(subset)

print('\nVerification:')
print('Subset shape:', subset.shape)
print('Subset columns:', subset.columns.tolist())
```

## Milestone Summary
- You selected single and multiple columns by name.
- You selected rows by position using `iloc`.
- You selected rows by label using `loc`.
- You combined row and column selection in clear, intentional expressions.

## Common Mistakes to Avoid
- Mixing up `iloc` (position) and `loc` (label)
- Forgetting that `iloc` slicing is stop-exclusive
- Forgetting that `loc` label slicing is typically stop-inclusive
- Using unclear chained selection instead of one explicit `loc`/`iloc` expression
- Not checking the output shape/columns after selection
