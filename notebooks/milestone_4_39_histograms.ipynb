# Milestone 4.39: Visualizing Data Distributions Using Histograms

**Objective:** Use histograms to visualize and compare numeric distributions so patterns become visible beyond summary statistics.

**Key Learning:**
- Histograms show how values are distributed across ranges
- Bins group numeric values and frequencies show counts per bin
- Shape reveals skewness, spread, and possible outliers
- Multi-column histogram comparison improves EDA insight

---

## Why Histograms Matter

Common beginner mistakes:
- Relying only on mean/median without visual checks
- Missing skewed or uneven distributions
- Overlooking unusual values that affect analysis
- Confusing histograms with bar charts

Professional habit: pair numeric summaries with histogram inspection.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

print('Pandas version:', pd.__version__)
```

---

## Part 1: Load Data and Select Numeric Columns

```python
sales_df = pd.read_csv('../data/raw/sales_data.csv')
numeric_cols = sales_df.select_dtypes(include='number').columns.tolist()

print('Numeric columns:', numeric_cols)
print('\nData preview:')
print(sales_df.head())
```

Histogram basics:
- `bins` split the numeric range into intervals
- Bar height represents frequency (count of values in each bin)
- Histograms describe distribution of continuous values

---

## Part 2: Create a Histogram for a Single Column

```python
plt.figure(figsize=(7, 4))
plt.hist(sales_df['Quantity'], bins=6, edgecolor='black')
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
print('Quantity summary statistics:')
print(sales_df['Quantity'].describe())
```

Interpretation prompts:
- Where are most observations concentrated?
- Is the distribution symmetric or skewed?
- Are there gaps or isolated high-frequency bins?

---

## Part 3: Read Shape and Skewness

```python
quantity_skew = sales_df['Quantity'].skew()
print('Quantity skewness:', quantity_skew)

if quantity_skew > 0:
    print('Interpretation: Right-skewed distribution (longer right tail).')
elif quantity_skew < 0:
    print('Interpretation: Left-skewed distribution (longer left tail).')
else:
    print('Interpretation: Roughly symmetric distribution.')
```

Skewness is a signal, not a final conclusion. Always inspect histogram shape directly.

---

## Part 4: Compare Histograms Across Columns

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].hist(sales_df['Quantity'], bins=6, edgecolor='black')
axes[0].set_title('Quantity Distribution')
axes[0].set_xlabel('Quantity')
axes[0].set_ylabel('Frequency')

axes[1].hist(sales_df['Price'], bins=6, edgecolor='black')
axes[1].set_title('Price Distribution')
axes[1].set_xlabel('Price')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
comparison_stats = sales_df[['Quantity', 'Price']].describe().T
comparison_stats['range'] = comparison_stats['max'] - comparison_stats['min']

print('Comparison statistics (to support histogram reading):')
print(comparison_stats[['mean', '50%', 'std', 'min', 'max', 'range']])
```

Comparison prompts:
- Which column has wider spread visually?
- Which appears more skewed?
- Do histogram impressions align with `std` and `range`?

---

## Part 5: Optional Multi-Column Histogram Grid

```python
# Quick way to view all numeric distributions together
sales_df[numeric_cols].hist(bins=6, figsize=(12, 4), edgecolor='black')
plt.suptitle('All Numeric Column Histograms', y=1.03)
plt.tight_layout()
plt.show()
```
