This notebook demonstrates how to use the project structure effectively.

## Step 1: Import Libraries and Custom Functions

We'll import pandas and our custom data processing functions from `src/`.

````python
import pandas as pd
import matplotlib.pyplot as plt
import sys

# Add src/ to path to import custom modules
sys.path.append('../src')
from data_processing import clean_sales_data, calculate_revenue, get_top_customers

print("‚úì Libraries and custom functions imported successfully!")
````

## Step 2: Load Raw Data

Load the original data from `data/raw/` folder.

````python
# Load raw data (read-only, never modify!)
df_raw = pd.read_csv('../data/raw/sample_sales.csv')

print(f"Raw data shape: {df_raw.shape}")
print("\nFirst few rows:")
df_raw.head()
````

## Step 3: Clean the Data

Use our custom function from `src/data_processing.py` to clean the data.

````python
# Clean the data using our custom function
df_cleaned = clean_sales_data(df_raw)

print(f"Cleaned data shape: {df_cleaned.shape}")
print(f"\nData quality check:")
print(f"  - Missing values: {df_cleaned.isnull().sum().sum()}")
print(f"  - Duplicates: {df_cleaned.duplicated().sum()}")
````

## Step 4: Save Processed Data

Save the cleaned data to `data/processed/` for future use.

````python
# Save cleaned data to processed folder
df_cleaned.to_csv('../data/processed/sample_sales_cleaned.csv', index=False)

print("‚úì Cleaned data saved to data/processed/sample_sales_cleaned.csv")
````

## Step 5: Analyze the Data

Perform analysis using our custom functions.

````python
# Calculate total revenue
total_revenue = calculate_revenue(df_cleaned)
print(f"Total Revenue: ${total_revenue:,.2f}")

# Get top customers
print("\nTop 5 Customers by Purchase Amount:")
top_customers = get_top_customers(df_cleaned, n=5)
top_customers
````

## Step 6: Create Visualization

Generate a plot and save it to `outputs/figures/`.

````python
# Create visualization
plt.figure(figsize=(10, 6))
plt.bar(top_customers['customer_id'].astype(str), top_customers['total_amount'])
plt.xlabel('Customer ID')
plt.ylabel('Total Purchase Amount ($)')
plt.title('Top 5 Customers by Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()

# Save figure to outputs/figures/
plt.savefig('../outputs/figures/top_customers.png', dpi=300, bbox_inches='tight')
print("‚úì Figure saved to outputs/figures/top_customers.png")

plt.show()
````

## Step 7: Generate Report

Create a summary report and save to `outputs/reports/`.

````python
# Generate summary statistics
summary_stats = df_cleaned.describe()

# Save to outputs/reports/
summary_stats.to_csv('../outputs/reports/sales_summary_statistics.csv')
print("‚úì Summary statistics saved to outputs/reports/sales_summary_statistics.csv")

print("\nSummary Statistics:")
summary_stats
````
